```python
from pylab import * 
import netCDF4

name = "CDP_hor_19930801_20170731_with_htn_ref_ts_ref.nc"

hor=netCDF4.Dataset(name)
time = hor.variables['time']
date=netCDF4.num2date(time[:], time.units)
```

```python
%matplotlib notebook
def read(var):
    variable=hor.variables[var]
    plot(date,variable,'+')
    show()
```
